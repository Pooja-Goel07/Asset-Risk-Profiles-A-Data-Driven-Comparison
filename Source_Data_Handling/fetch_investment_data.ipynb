```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

url = "https://finance.yahoo.com/quote/BTC-USD/history/?period1=1609459200&period2=1743465600&frequency=1mo"  # Replace with your actual URL

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Get the page
page = requests.get(url, headers=headers)
soup = BeautifulSoup(page.text, 'html.parser')

# Find the div by its class
table_div = soup.find('div', class_='table-container yf-1jecxey')

# Print the content inside that div
table = table_div.find('table')
df = pd.read_html(StringIO(str(table)))[0]

# Rename the 6th column (index 5) and the column with "Close Close price adjusted for splits"
df.columns.values[4] = 'Close'
df.columns.values[5] = 'Adj Close'

# Display the DataFrame with the new column names
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(df)
```

```python
df.to_csv('BTC-USD_data.csv', index=False)
```

```python
import requests
import pandas as pd
from io import StringIO

url = "https://finance.yahoo.com/quote/TSLA/history/?period1=1609459200&period2=1743552000&frequency=1mo"
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Get the page
response = requests.get(url, headers=headers)

# Wrap the HTML response in StringIO to avoid the warning
tables = pd.read_html(StringIO(response.text))


# The first table contains the historical data
df = tables[0]

# Optional: remove non-price rows like "Dividend" or "Stock Split"
df = df[~df['Open'].astype(str).str.contains("Dividend|Split", na=False)]

df.columns.values[4] = 'Close'
df.columns.values[5] = 'Adj Close'


# Display full table
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(df)
```

```python
df.to_csv('TSLA_data.csv', index=False)
```
