```python
import requests
import pandas as pd
from io import StringIO

url = "https://finance.yahoo.com/quote/TSLA/history/?period1=1609459200&period2=1743465600&frequency=1mo"
headers = {
    "User-Agent": "Mozilla/5.0"
}

# Get the page
response = requests.get(url, headers=headers)

# Wrap the HTML response in StringIO to avoid the warning
tables = pd.read_html(StringIO(response.text))

# The first table contains the historical data
df = tables[0]

# Optional: remove non-price rows like "Dividend" or "Stock Split"
df = df[~df['Open'].astype(str).str.contains("Dividend|Split", na=False)]

# Display full table
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(df)
```

```python
df.to_csv('TSLAdata.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

# Yahoo Finance historical data URL (example: AGG)
url = "https://finance.yahoo.com/quote/AGG/history/?period1=1609459200&period2=1743465600&frequency=1mo"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Request the page
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Find the div containing the table
table_div = soup.find('div', class_='table-container')
table = table_div.find('table')

# Read the table as-is (without dropping anything)
df = pd.read_html(StringIO(str(table)))[0]

# Display all rows and columns
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(df)
```

```python
df.to_csv('AGG_data.csv', index=False)
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from io import StringIO

# Yahoo Finance historical data for BIL
url = "https://finance.yahoo.com/quote/BIL/history/?period1=1609459200&period2=1743552000&frequency=1mo"

headers = {
    "User-Agent": "Mozilla/5.0"
}

# Request page content
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

# Locate the table
table_div = soup.find('div', class_='table-container')
table = table_div.find('table')

# Read HTML table as-is
df = pd.read_html(StringIO(str(table)))[0]

# Show full table without dropping/modifying anything
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
display(df)
```

```python
df.to_csv('BIL_data.csv', index=False)
```
