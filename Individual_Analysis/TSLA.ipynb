## 1. Tesla (TSLA) Analysis - High Risk (Growth Stock)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from matplotlib.ticker import FuncFormatter # Import FuncFormatter

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 4)
warnings.filterwarnings('ignore') 
```

### Load Processed Data

```python
input_data_folder = '../Cleaned_Exports'

adj_close_filename = 'aligned_adj_close_prices.csv'
returns_filename = 'calculated_monthly_returns.csv'

adj_close_filepath = os.path.join(input_data_folder, adj_close_filename)
returns_filepath = os.path.join(input_data_folder, returns_filename)

adj_close_df = pd.read_csv(adj_close_filepath, index_col='Date', parse_dates=True)
monthly_returns = pd.read_csv(returns_filepath, index_col='Date', parse_dates=True)

print("\nData loaded successfully.")
print(f"Price data shape: {adj_close_df.shape}")
print(f"Return data shape: {monthly_returns.shape}")
```

#### Calculate Rolling Volatility

```python
# Calculate rolling volatility based on loaded monthly returns
if 'monthly_returns' in locals() and not monthly_returns.empty:
    rolling_window = 12 # Define rolling window size
    rolling_std = monthly_returns.rolling(window=rolling_window).std()
    annualized_rolling_volatility = rolling_std * np.sqrt(12)
    print(f"\nCalculated {rolling_window}-Month Annualized Rolling Volatility.")
else:
    print("\nSkipping rolling volatility calculation as monthly returns data is not available.")
    # Assign an empty DataFrame or None if returns weren't loaded
    annualized_rolling_volatility = pd.DataFrame()
```

### **TSLA:** Price Trend

```python
asset_ticker = 'TSLA'
print(f"--- {asset_ticker} Price Analysis ---")

if asset_ticker in adj_close_df.columns:
    plt.figure() # Use default figsize set earlier or specify here e.g., figsize=(12,4)
    adj_close_df[asset_ticker].plot()
    plt.title(f'{asset_ticker} Adjusted Close Price Trend')
    plt.ylabel('Price (USD)')
    plt.xlabel('Date')
    plt.grid(True, ls='--', linewidth=0.5)
    plt.show()
else:
    print(f"{asset_ticker} price data not found.")
```


### **Interpretation:** TSLA Price Trend

The price trend for TSLA from early 2021 to early 2025 clearly demonstrates the characteristics of a volatile growth stock. Key observations include:

1.  **High Volatility:** The line chart shows significant price fluctuations throughout the period, with sharp peaks and deep troughs rather than a smooth progression.
2.  **Major Price Swings:**
    *   The stock reached notable highs around late 2021/early 2022 (approaching 400 dollars) and again in late 2024/early 2025 (reaching above 400 dollars).
    *   Conversely, it experienced a major drawdown, hitting a low point below 150 dollars sometime in late 2022 or early 2023.
3.  **Lack of Consistent Trend:** While exhibiting large intermediate movements, the price at the end of the observed period (around 250 dollars in early 2025) is relatively close to the starting price (around 265 dollars in early 2021), indicating significant volatility without a strong overall upward or downward trend across this specific timeframe.

Overall, this visual representation strongly confirms TSLA's classification as a high-risk asset, subject to substantial gains and losses over relatively short periods.



### **TSLA** - Return Analysis

```python
asset_ticker = 'TSLA'
print(f"--- {asset_ticker} Return Analysis ---")

if asset_ticker in monthly_returns.columns:
    # Distribution Plot
    plt.figure()
    sns.histplot(monthly_returns[asset_ticker], kde=True, bins=20)
    plt.title(f'{asset_ticker} Monthly Return Distribution')
    plt.xlabel('Monthly Return')
    plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:.1%}'.format(x)))
    plt.show()

    # Descriptive Statistics
    print(f"\n{asset_ticker} Monthly Return Statistics:")
    try:
        print(monthly_returns[asset_ticker].describe().to_string(float_format='{:.4f}'.format))
    except Exception as e:
        print(f"Could not calculate describe() stats for {asset_ticker}: {e}")
else:
    print(f"{asset_ticker} return data not found.")
```

### **Interpretation:** TSLA Return Analysis

This analysis examines the month-over-month percentage changes for TSLA.

1.  **Distribution Shape and Spread:** The histogram shows a wide distribution of monthly returns, visually confirming significant volatility. Returns range from substantial losses to substantial gains, typical of a high-risk growth stock. The distribution appears roughly centered around zero, though slightly skewed by large positive and negative outcomes.
2.  **Statistical Summary:**
    *   **Mean vs. Median:** The average monthly return (`mean`) is positive at 1.41%. However, the median monthly return (`50%`) is slightly negative at -1.12%. This suggests that while the average outcome was positive, slightly more than half the months experienced negative returns, and the positive average was likely driven by some months with particularly large gains.
    *   **Volatility (Standard Deviation):** The monthly standard deviation (`std`) is 17.80%. This is a high value, quantitatively confirming the large dispersion of returns seen in the histogram and indicating significant month-to-month risk.
    *   **Range (Min/Max):** The most extreme monthly outcomes observed were a loss of 36.73% (`min`) and a gain of 43.65% (`max`). This wide range further underscores the high volatility and the potential for large price swings in either direction.

Collectively, the wide distribution, high standard deviation, and extreme min/max values strongly support the classification of TSLA as a high-volatility, high-risk investment during this period.

### **TSLA** - Rolling Volatility Analysis

```python
asset_ticker = 'TSLA'
print(f"--- {asset_ticker} Rolling Volatility Analysis ---")

if 'annualized_rolling_volatility' in locals() and not annualized_rolling_volatility.empty and asset_ticker in annualized_rolling_volatility.columns:
    plt.figure()
    annualized_rolling_volatility[asset_ticker].plot()
    plt.title(f'{asset_ticker} {rolling_window}-Month Rolling Annualized Volatility')
    plt.ylabel('Annualized Volatility')
    plt.xlabel('Date')
    plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.1%}'.format(y)))
    plt.grid(True, ls='--', linewidth=0.5)
    plt.show()
else:
    print(f"{asset_ticker} rolling volatility data not found or could not be calculated.")
```

### **Interpretation:** TSLA Rolling Volatility

This plot illustrates how TSLA's estimated annualized volatility changed over time, based on a 12-month rolling window of its monthly returns. It provides a dynamic view of risk, complementing the overall average volatility calculated earlier (~61.7%).

1.  **Consistently High Volatility:** The rolling volatility remained significantly elevated throughout the entire period shown (starting 2022), generally fluctuating between approximately 45% and 85%. This confirms that TSLA wasn't just volatile on average, but experienced persistently high levels of risk.
2.  **Significant Fluctuations in Risk:** Risk was not static. There was a clear upward trend in volatility through 2022, peaking near 85% in the first half of 2023. This indicates a period of particularly heightened uncertainty or larger price swings.
3.  **Period of Decreasing Volatility:** Following the 2023 peak, volatility generally declined, reaching a relative low point below 45% in early 2024, suggesting a comparatively calmer (though still high) period for the stock.
4.  **Recent Uptick:** Towards the end of the observed period (late 2024 into 2025), volatility appears to be trending upwards again, moving back above the 60% mark.

This dynamic view shows that while TSLA is inherently a high-volatility asset, the *degree* of that volatility varied considerably over the analysis timeframe.

## **Conclusion for TSLA**

Tesla (TSLA) exhibited patterns characteristic of a highly volatile growth stock during the 2021-2025 period. Its price experienced significant swings, including peaks near 400 dollars and troughs below 150 dollars, without establishing a clear overall trend from start to finish. This visual volatility was quantified by a wide monthly return distribution (ranging from -37% to +44%) and a high monthly standard deviation (around 18%). Furthermore, its 12-month rolling volatility remained elevated, mostly fluctuating between 45% and 85%, confirming its classification as a **high-risk** asset. Despite this risk and a negative median monthly return, TSLA generated a positive average monthly return (~1.4%), indicating it was **beneficial** in terms of average capital appreciation during this specific timeframe, albeit accompanied by substantial risk.
