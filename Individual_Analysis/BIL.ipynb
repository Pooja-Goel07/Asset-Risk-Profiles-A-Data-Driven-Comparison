# 4. SPDR Bloomberg 1-3 Month T-Bill ETF (**BIL**) Analysis - Low Risk (T-Bill Fund)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
from matplotlib.ticker import FuncFormatter # Import FuncFormatter

sns.set(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 4)
warnings.filterwarnings('ignore') 
```

### Load Processed Data

```python
input_data_folder = '../Cleaned_Exports'

adj_close_filename = 'aligned_adj_close_prices.csv'
returns_filename = 'calculated_monthly_returns.csv'

adj_close_filepath = os.path.join(input_data_folder, adj_close_filename)
returns_filepath = os.path.join(input_data_folder, returns_filename)

adj_close_df = pd.read_csv(adj_close_filepath, index_col='Date', parse_dates=True)
monthly_returns = pd.read_csv(returns_filepath, index_col='Date', parse_dates=True)

print("\nData loaded successfully.")
print(f"Price data shape: {adj_close_df.shape}")
print(f"Return data shape: {monthly_returns.shape}")
```

### Calculate Rolling Volatility

```python
# Calculate rolling volatility based on loaded monthly returns
if 'monthly_returns' in locals() and not monthly_returns.empty:
    rolling_window = 12 # Define rolling window size
    rolling_std = monthly_returns.rolling(window=rolling_window).std()
    annualized_rolling_volatility = rolling_std * np.sqrt(12)
    print(f"\nCalculated {rolling_window}-Month Annualized Rolling Volatility.")
else:
    print("\nSkipping rolling volatility calculation as monthly returns data is not available.")
    # Assign an empty DataFrame or None if returns weren't loaded
    annualized_rolling_volatility = pd.DataFrame()
```

## **BIL** - Price Trend

```python
asset_ticker = 'BIL'
print(f"--- {asset_ticker} Price Analysis ---")

if asset_ticker in adj_close_df.columns:
    plt.figure()
    adj_close_df[asset_ticker].plot()
    plt.title(f'{asset_ticker} Adjusted Close Price Trend')
    plt.ylabel('Price (USD)')
    plt.xlabel('Date')
    # Focus y-axis limits to see small changes typical for BIL
    min_price = adj_close_df[asset_ticker].min()
    max_price = adj_close_df[asset_ticker].max()
    # Add a small buffer, ensure min is not equal to max
    y_buffer = (max_price - min_price) * 0.1 if max_price > min_price else 0.1
    plt.ylim(min_price - y_buffer, max_price + y_buffer)
    plt.grid(True, ls='--', linewidth=0.5)
    plt.show()
else:
    print(f"{asset_ticker} price data not found.")
```


### **Interpretation**: BIL Price Trend

The price trend for the BIL ETF (1-3 Month T-Bills) clearly reflects its function as an ultra-low-risk, cash-equivalent investment.

1.  **Exceptional Stability:** For the initial part of the period (2021 through 2022), the price is remarkably stable, hovering just below 81 dollars with minimal fluctuation. This signifies extremely low price volatility.
2.  **Steady Accrual:** Beginning in late 2022/early 2023, the price exhibits a smooth and steady upward trend, climbing consistently to over 91 dollars by early 2025. This upward movement primarily reflects the accrual of interest from the underlying short-term Treasury bills.
3.  **Interest Rate Sensitivity (Yield):** The noticeable steepening of the upward trend from 2023 onwards likely corresponds to the period of rising short-term interest rates set by the Federal Reserve. As yields on new T-bills increased, the ETF's price reflected this higher rate of accrual.
4.  **Capital Preservation Focus:** The chart lacks the sharp peaks and troughs seen in TSLA or BTC-USD, demonstrating its focus on capital preservation rather than capital appreciation through price swings.

Overall, the BIL price trend confirms its role as a very low-risk asset whose value changes primarily due to the slow and steady accrual of interest from short-term government debt, making it a suitable proxy for the risk-free rate in this analysis.

## **BIL** - Return Analysis

```python
asset_ticker = 'BIL'
print(f"--- {asset_ticker} Return Analysis ---")

if asset_ticker in monthly_returns.columns:
    # Distribution Plot
    plt.figure()
    sns.histplot(monthly_returns[asset_ticker], kde=True, bins=15) # Fewer bins suitable for tight data
    plt.title(f'{asset_ticker} Monthly Return Distribution')
    plt.xlabel('Monthly Return')
    # Use more precision on axis for potentially small returns
    plt.gca().xaxis.set_major_formatter(FuncFormatter(lambda x, _: '{:.2%}'.format(x)))
    plt.show()

    # Descriptive Statistics
    print(f"\n{asset_ticker} Monthly Return Statistics:")
    try:
        # Use more precision for describe output
        print(monthly_returns[asset_ticker].describe().to_string(float_format='{:.6f}'.format))
    except Exception as e:
        print(f"Could not calculate describe() stats for {asset_ticker}: {e}")
else:
    print(f"{asset_ticker} return data not found.")
```


### **Interpretation:** BIL Return Analysis

This analysis reviews the monthly percentage returns for the BIL T-Bill ETF.

1.  **Distribution Shape and Spread:** The histogram shows an extremely narrow distribution, heavily concentrated at the low end of the positive return spectrum. The vast majority of returns fall between 0.00% and approximately 0.50%. This visually confirms the asset's exceptionally low volatility.
2.  **Statistical Summary:**
    *   **Mean and Median:** The average monthly return (`mean`) was 0.2361% and the median return (`50%`) was 0.1359%. Both are small positive values, reflecting the modest yields from short-term T-bills during the period.
    *   **Volatility (Standard Deviation):** The monthly standard deviation (`std`) is incredibly low at 0.3026%. This is orders of magnitude smaller than the volatility observed for AGG, TSLA, or BTC-USD, quantifying its minimal risk profile.
    *   **Range (Min/Max):** The returns are tightly clustered. The lowest recorded monthly return (`min`) was -0.0469% (a very small loss, possibly due to expense ratios or minor fluctuations), and the highest monthly return (`max`) was 1.3490%. This very narrow range between the extremes further emphasizes the low risk.

The return distribution and statistics clearly demonstrate that BIL provided consistent, small positive returns with minimal fluctuation, aligning perfectly with its role as a low-risk cash equivalent or risk-free proxy.


## **BIL** - Rolling Volatility Analysis

```python
asset_ticker = 'BIL'
print(f"--- {asset_ticker} Rolling Volatility Analysis ---")

if 'annualized_rolling_volatility' in locals() and not annualized_rolling_volatility.empty and asset_ticker in annualized_rolling_volatility.columns:
    plt.figure()
    annualized_rolling_volatility[asset_ticker].plot()
    plt.title(f'{asset_ticker} {rolling_window}-Month Rolling Annualized Volatility')
    plt.ylabel('Annualized Volatility')
    plt.xlabel('Date')
    # Focus y-axis because BIL volatility is very low
    max_vol = annualized_rolling_volatility[asset_ticker].max()
    if pd.notna(max_vol) and max_vol < 0.03: # If max vol is valid and low
         plt.ylim(0, max_vol * 1.5 if pd.notna(max_vol) else 0.01) # Adjust upper limit safely
    plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: '{:.2%}'.format(y)))
    plt.grid(True, ls='--', linewidth=0.5)
    plt.show()
else:
    print(f"{asset_ticker} rolling volatility data not found or could not be calculated.")
```


### **Interpretation:** BIL Rolling Volatility

This plot shows the dynamic risk profile of the BIL T-Bill ETF using a 12-month rolling annualized volatility calculation.

1.  **Note on Start Date:** The plot begins in early 2022, as 12 months of prior data are required for the first calculation point.
2.  **Extremely Low Volatility Level:** The most striking feature is the scale of the y-axis, which only goes up to 1.75%. The rolling volatility for BIL remained exceptionally low throughout the entire period, consistently staying below 1.25%. This is orders of magnitude lower than the other assets analyzed, confirming its ultra-low-risk nature.
3.  **Gradual Increase:** Despite the low absolute level, there was a noticeable pattern. Volatility started near zero in early 2022 and gradually increased through 2023, reaching a slightly higher plateau around 0.85%. It then stepped up again near the beginning of 2024 to around 1.20% before slightly decreasing towards the end of the period.
4.  **Context:** While still minimal, this slight increase in measured volatility likely reflects minor fluctuations around the changing short-term interest rate environment during 2022-2024. However, the key takeaway is the persistently negligible level of risk.

This chart visually reinforces BIL's role as a highly stable, near cash-equivalent investment with minimal price risk over time.
